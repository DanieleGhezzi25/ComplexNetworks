# Connected Components
In this software are called components

```r
set.seed(1234)
N <- 1000
g <- erdos.renyi.game(N, p=0.1/N, directed=F)
lay <- layout_with_fr(g)
plot(g, vertex.color="steelblue", vertex.label=NA, edge.curved=0.2, layout=lay, vertex.size=4*log10(degree(g)), edge.color="gray80", vertex.frame.color=NA)
```

```r
clu <- components(g)
clu
```

```r
ggplot( data.frame(s=clu$csize), aes(s) ) + theme_bw() + geom_histogram() 
# geom_histogram is a way to print an histogram (all the + are layers above the figure !)
```

We want to see the size of the greatest connected component as we increase p (wiring probability)

```r
res <- data.frame()
for(p in seq(0.1/N, 2*log(N)/N, 0.02/N)){
    g <- erdos.renyi.game(N, p=p, directed=F)
    clu <- components(g)
    res <- rbind(res, data.frame(p=p, n=clu$no, lcc=max(clu$csize)/N))
}
res
ggplot(res, aes(p,n)) + theme_bw() + geom_line() 
ggplot(res, aes(p,lcc)) + theme_bw() + geom_line() + scale_x_log10() + geom_vline(xintercept = 1/N, linetype="dashed", color="red")
```

We do not have zero before the critical threshold as we do not have the thermodynamic limit as in theory. We should infact repeat the toy model multiple times and then take the averages
